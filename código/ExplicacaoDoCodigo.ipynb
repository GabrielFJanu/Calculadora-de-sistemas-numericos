***verificar_valido:***
Função para verificar se a entrada é válida

---



O usuário não pode digitar um input inválido, pois isso apresentará erro no código. Exemplo: um número com algarismos incompatíveis com a base, números com dois pontos... etc

Por isso, existe esta função "verificar_valido". Nela, as entradas vão ser 1 string e um inteiro (o número e sua respectiva base) e a saída vai ser um booleano indentificando se aquele número é válido (True) ou não (False).

A primeira etapa do código é contruir um vetor com todos os símbolos possíveis da base b. Depois, compara se todos os caracteres do número x pertencem ao vetor. Outra coisa é analisar se o número tem mais de um ponto.

A única forma de retornar True (valor válido) é se todos os algarismos do número pertencerem a base e ele não tiver mais de um ponto.

```python
"""verificando se a entrada é válida"""
def verificar_valido(x,b): # string --> bool
  #formando o vetor com os símbolos das bases
  base = []
  if b <=10:
    for i in range(b):
      base.append(str(i))
  elif b>10:
    for i in range(10):
      base.append(str(i))
    for i in range(b-10):
      base.append(chr(65+i))
  #validando se todos os caracteres pertencem à base e se possui mais de um "."
  pontos = 0
  for i in range(len(x)):
    flag = False
    for j in range(len(base)):
      if x[i] == base[j] or x[i]==".":
        flag = True
    if not flag:
      return False
    if x[i] == ".":
      pontos+=1
  if pontos >1:
    return False
  return True
```

***char_to_int:***
Função para transformar os algarismos para o seu respectivo valor inteiro na base 10


---


Essa função tem como entrada um vetor char e como saída um vetor de inteiros

***int_to_char:***
Função para transformar números inteiros na base 10 para seus respectivos símbolos


---


Essa função tem como entrada um vetor de inteiros e saída um vetor de char

```python
"""transformando os símbolos do vetor em números equivalentes na base 10"""
def char_to_int(v): # vetor char --> vetor int
  for i in range(len(v)):
    if v[i]>= "A" and v[i]<="Z":
      v[i] = ord(v[i])-55
    else:
      v[i] = int(v[i])
  return v

"""transformando os números na base 10 em símbolos equivalentes"""
def int_to_char(v): # vetor int --> vetor char
  for i in range(len(v)):
    if v[i] >=10:
      v[i] = chr(v[i]+55)
    else:
      v[i] = str(v[i])
  return v
```

***b10_to_bB:***
Função que converte um número da base 10 para a base b


> Usado somente em mudar_base




---

Função tem como entrada o número (string), a sua base (int) e a precisão de casas calculadas (int). É preciso esse último parâmetro para o algoritmo não calcular infinitas casas, pois alguns números dão dízimas periódicas.

Parte 1:

1.   transformar x (string) em real (float).
1.   Separar sua parte inteira da fracionaria.



Parte 2.1:
1.   Converter a parte inteira por meio do algoritmo de euclides (divisão com resto). ***COMO ENSINADO EM SALA DE AULA.***
1.   Armazenar o resultado no vetor "resultado_int".
1.   Transformar o vetor "resultado_int" em um vetor char (caso inteiro == 0, então resultado_int = ["0"]. Isso para evitar que a função retorne vazio caso x = "0").

Parte 2.2:
1.   Converter a parte fracionada por meio do algoritmo da multiplicação ***COMO ENSINADO EM SALA DE AULA.***
1.   Armarzenar o resultado no vetor "resultado_frac".
1.   Transformar o vetor "resultado_frac" em um vetor char

Parte 3:
1.   Junção dos resultados em um string, retorna o número já convertido na base b


```python
"""convertendo um número da base 10 para a base B"""
def b10_to_bB(x, b, casas_calculadas):
  x = float(x)
  inteiro, fracionario = int(x), x - int(x)
  # Convertendo a parte inteira
  resultado_int = []
  while inteiro > 0:
    resultado_int.insert(0, inteiro % b)
    inteiro //= b
  resultado_int = int_to_char(resultado_int) or ["0"]
  # Convertendo a parte fracionária
  resultado_frac = []
  for i in range(casas_calculadas):
    fracionario *= b
    digito = int(fracionario)
    resultado_frac.append(digito)
    fracionario -= digito
    if fracionario == 0:
      break
  resultado_frac = int_to_char(resultado_frac)
  # Construindo o resultado final
  if resultado_frac:
    return "".join(resultado_int) + "." + "".join(resultado_frac)
  else:
    return "".join(resultado_int)
```

***somar_mesma_base:***
Função que soma dois números de bases iguais. A operação acontece na própria base.


---

A função tem como entrada dois números (strings) e sua base (int).

Parte 1:
1.   transformar x e y em vetores char.

Parte 2:
1.   calcular quantas casas decimais e inteiras cada número tem.
1.   tirar o char "." dos vetores x e y para não atrapalhar futuro cálculo.

Parte 3:
1.   transformar os vetores char em vetores inteiros.

Parte 4:
1.   colocando os vetores com o mesmo tamanho para o futuro cálculo. Como o cálculo da soma é feito coordenada por coordenada, essa etapa é importante para não misturar unidade com dezena, centena... etc.
Exemplo: x = "1003" e y = "1.13" --> x = "1003.00" e y = "0001.13"

Parte 5:
1.   Adicionando mais um 0 à esquerda em x e y para possível "vai um" no final.

Parte 6:
1.   Cálculo da soma coordenada por coordenada. ***COMO ENSINADO EM SALA DE AULA***.
1.   Armazenamento do resultado no vetor resultado.

Parte 7:
1.   transformando o vetor resultado (inteiro) para char.

Parte 8:
1.   Adicionando o "." no vetor resultado para separar parte inteira da fracionaria.

Parte 9:
1.   tirando possível 0 à esquerda (inteiros) e à direita (fracionados) e possível "." do vetor resultado (etapa desnecessária, só para deixar o resultado mais "limpo").
Exemplo: 0010.00 --> 10

Parte 10:
1.   retornando resultado, transformando em string.




```python
"""somar números da mesma base"""
def somar_mesma_base(x,y,b):
  x = list(x)
  y = list(y)
  #calculando quantas casas inteiras e decimais cada número tem
  if "." in x:
    casas_frac_x = len(x) - x.index(".") - 1
    x.remove(".")
  else:
    casas_frac_x = 0
  if "." in y:
    casas_frac_y = len(y) - y.index(".") - 1
    y.remove(".")
  else:
    casas_frac_y = 0
  casas_int_x = len(x)-casas_frac_x
  casas_int_y = len(y)-casas_frac_y
  # tranformando os símbolos em números equivalentes na base 10
  char_to_int(x)
  char_to_int(y)
  #colocando os vetores com o mesmo número de casas
  if casas_int_x>casas_int_y:
    for i in range(casas_int_x-casas_int_y):
      y.insert(0,0)
  else:
    for i in range(casas_int_y-casas_int_x):
      x.insert(0,0)
  if casas_frac_x>casas_frac_y:
    for i in range(casas_frac_x-casas_frac_y):
      y.append(0)
  else:
    for i in range(casas_frac_y-casas_frac_x):
      x.append(0)
  #adicionando mais uma casa para possível "vai um"
  x.insert(0,0)
  y.insert(0,0)
  #cálculo
  vai_um = 0
  resultado = []
  for i in range(len(x)-1,-1,-1):
    resultante = x[i]+y[i]+vai_um
    vai_um = 0
    if resultante>=b:
      vai_um+=1
      resultante-=b
    resultado.insert(0,resultante)
  #transformando números na base 10 para símbolos equivalentes
  int_to_char(resultado)
  #adicionando o "."
  resultado.insert(max(casas_int_x,casas_int_y)+1,".")
  #tirando possível 0 à esquerda (inteiros) e à direita (fracionados) e possível "."
  while resultado[0] == "0" and resultado[1] != ".":
    del resultado[0]
  while resultado[len(resultado)-1] == "0":
    del resultado[len(resultado)-1]
  if resultado[len(resultado)-1] == ".":
    del resultado[len(resultado)-1]
  #transformando resultado em string
  resultado = "".join(resultado)
  return resultado
```

***multiplicar_mesma_base:***
Função para multiplicar dois números de bases iguais. A operação acontece na própria base.


---

A função tem como entrada dois números (strings) e sua base (int).

Parte 1:
1.   transformar x e y em vetores char.

Parte 2:
1.   calcular quantas casas decimais e inteiras cada número tem.
1.   tirar o char "." dos vetores x e y para não atrapalhar futuro cálculo.

Parte 3:
1.   transformar os vetores char em vetores inteiros.

Parte 4:
1.   Adicionando mais um 0 à esquerda em x para possível "vai um" no cálculo da multiplicação.


Parte 5:
 Cálculo da multiplicação. ***COMO ENSINADO EM SALA DE AULA***.


Parte 5.1:
1.   criação da matriz resultado, com cada linha (resultado[i]) representando o futuro produto do número x com o i-ésimo algarismo de y da direita para esquerda.
1.   adicionar os zeros à direita necessários nos vetores resultado[i] para a etapa da soma.
1.   etapa da multiplicação.
1.   armazenar os produtos do número x com o i-ésimo algarismo de y (da direita para esquerda) nos vetores resultado[i]
1.   transformar os vetores resultado[i] (inteiro) em char
1.   transformar os vetores resultado[i] (char) em string

Parte 5.2:
1.   etapa da soma. Somar todos os strings resultado[i] e armazenar na string soma

Parte 6:
1.   transformar a string soma em um vetor char
1.   adicionar o char "." para separar a parte inteira da fracionária (há casos que é adicionado "0"s à esquerda no vetor soma para colocar o "." no lugar correto)

Parte 7:
1.   tirando possível 0 à direita e possível "." (etapa desnecessária, só para deixar o resultado mais "limpo"). Exemplo: 10.00 --> 10

Parte 8:
1.   transformar o vetor soma para string
1.   retornar a string soma

```python
"""multiplicar dois números de bases iguais"""
def multiplicar_mesma_base(x,y,b):
  x = list(x)
  y = list(y)
  #calculando quantas casas decimais cada número tem
  if "." in x:
    casas_frac_x = len(x) - x.index(".") - 1
    x.remove(".")
  else:
    casas_frac_x = 0
  if "." in y:
    casas_frac_y = len(y) - y.index(".") - 1
    y.remove(".")
  else:
    casas_frac_y = 0
  # tranformando os símbolos em números equivalentes na base 10
  char_to_int(x)
  char_to_int(y)
  #adicionando mais uma casa para possível "vai um"
  x.insert(0,0)
  #cálculo multiplicação
  resultado = []
  for i in range(len(y)):
    resultado.append([])
    vai_um = 0
    for j in range(i):#botando casas necessarias nos vetores resultado[i]
      resultado[i].insert(0,0)
    for j in range(len(x)):
      produto = y[len(y)-1-i]*x[len(x)-1-j]+vai_um #int base 10
      resultado[i].insert(0,produto%b) #int
      vai_um = produto//b
    int_to_char(resultado[i]) #transformando int para char
    resultado[i] = "".join(resultado[i]) #transformando vetor char para str
  #cálculo soma
  soma = resultado[0]
  if len(resultado)>1:
    for i in range(len(resultado)-1):
      soma = somar_mesma_base(resultado[1+i],soma,b)
  #botando "."
  if (casas_frac_x+casas_frac_y) != 0:
    soma = list(soma)
    for i in range(1-(len(soma)-(casas_frac_x+casas_frac_y))):
      soma.insert(0,"0")
    soma.insert(len(soma)-(casas_frac_x+casas_frac_y),".")
    #tirando possível 0 à direita (fracionados) e possível "."
    while soma[len(soma)-1] == "0":
      del soma[len(soma)-1]
    if soma[len(soma)-1] == ".":
      del soma[len(soma)-1]
    soma = "".join(soma)
  return soma
```

***mudar_base:***
Função para converter um número da base 1 para a base 2 diretamente. Usando o método de horner.


---

A função tem como entrada um número (string), uma base inicial (int), uma base final (int) e a precisão de casas decimais (int) (máximo 16 casas devido as limitações de precisão python). É preciso esse último parâmetro para o algoritmo não calcular infinitas casas, pois alguns números dão dízimas periódicas.

Parte 1:
1.   aumentar a margem de casas calculadas no projeto para melhor precisão no cálculo (utilizando um modelo afim de aumento para não perder precisão com grandes casas_calculadas)

Parte 2:
1.   transformar o número x (string) em vetor char

Parte 3:
1.   calcular quantas casas decimais e inteiras de x
1.   tirar o char "." do vetor x para não atrapalhar futuro cálculo.

Parte 4:
1.   transformando o vetor x em um vetor inteiro

Parte 5: Cálculo método de horner. ***COMO ENSINADO EM SALA DE AULA***.
1.   converter o número base 1 para a base 2.
Exemplo: b1 = 9 e b2 = 7 --> 9(10) --> 12(7)
1.   converter o inverso multiplicativo base 1 para a base 2.
Exemplo: b1^-1 = 9^-1 e b2 = 7 --> 9^-1(10) --> 0.0530530...(7).
1.   etapa método de horner.
1.   armazenar resultado na string resultado

Parte 6:
1.   Tirando os algarismos duvidosos, ou seja, tirando a margem adicionada na parte 1.



```python
"""converter um número na base b1 para a base b2 diretamente (método de horner)"""
def mudar_base(x,b1,b2,casas_calculadas):
  casas_calculadas += 5 #margem de segurança para uma resposta precisa
  x = list(x)
  #calculando quantas casas decimais cada número tem
  if "." in x:
    casas_frac_x = len(x) - x.index(".") - 1
    x.remove(".")
  else:
    casas_frac_x = 0
  casas_int_x = len(x)-casas_frac_x
  #tranformando os símbolos em números equivalentes na base 10
  char_to_int(x)
  #cálculo
  base = b10_to_bB(b1,b2,casas_calculadas)
  inverso_base = b10_to_bB((1/b1),b2,casas_calculadas)
  resultado = b10_to_bB(x[0],b2,casas_calculadas)
  for i in range(len(x)-1):
    resultado = multiplicar_mesma_base(resultado,base,b2)
    resultado = somar_mesma_base(resultado,b10_to_bB(x[i+1],b2,casas_calculadas),b2)
  for i in range(casas_frac_x):
    resultado = multiplicar_mesma_base(resultado,inverso_base,b2)
  #tirando casas com algarismos duvidosos
  casas_calculadas -= 5 #tirando a margem
  resultado = list(resultado)
  if "." in resultado:
    casas_int_resultado = resultado.index(".")
  else:
    casas_int_resultado = len(resultado)
  while (casas_int_resultado+casas_calculadas+1) < len(resultado):
    del resultado[casas_int_resultado+casas_calculadas+1]
  resultado = "".join(resultado)
  #return
  return resultado
```

***multiplicar_diferentes_bases:*** Função que soma dois números em bases quaisquer.

Parte 3:
*   mudar "resposta" da base B1 para a base final, usando a função *mudar_base* e retorna-la

```python
"""somar dois números de bases diferentes"""
def somar_diferentes_bases(x,bx,y,by,bf,casas_calculadas): #(número 1, base do número 1, número 2, base do número 2, base final, precisão de casas)
  y = mudar_base(y,by,bx,casas_calculadas) #convertendo num 2 para a base 1
  resposta = somar_mesma_base(x,y,bx) #somando num 1 e num 2 na base 1
  resposta = mudar_base(resposta,bx,bf,casas_calculadas) #convertendo o resultado para a base final
  return resposta
"""multiplicar dois números de bases diferentes"""
def multiplicar_diferentes_bases(x,bx,y,by,bf,casas_calculadas): #(número 1, base do número 1, número 2, base do número 2, base final, precisão de casas)
  y = mudar_base(y,by,bx,casas_calculadas) #convertendo num 2 para a base 1
  resposta = multiplicar_mesma_base(x,y,bx) #multiplicando num 1 e num 2 na base 1
  resposta = mudar_base(resposta,bx,bf,casas_calculadas) #convertendo o resultado para a base final
  return resposta
```
