# Funções Secundárias


```python
"""verificando se a entrada é válida"""
def verificar_valido(x,b): # string --> bool
  #formando o vetor com os símbolos das bases
  base = []
  if b <=10:
    for i in range(b):
      base.append(str(i))
  elif b>10:
    for i in range(10):
      base.append(str(i))
    for i in range(b-10):
      base.append(chr(65+i))
  #validando se todos os caracteres pertencem à base e se possui mais de um "."
  pontos = 0
  for i in range(len(x)):
    flag = False
    for j in range(len(base)):
      if x[i] == base[j] or x[i]==".":
        flag = True
    if not flag:
      return False
    if x[i] == ".":
      pontos+=1
  if pontos >1:
    return False
  return True
```

```python
"""transformando os símbolos do vetor em números equivalentes na base 10"""
def char_to_int(v): # vetor char --> vetor int
  for i in range(len(v)):
    if v[i]>= "A" and v[i]<="Z":
      v[i] = ord(v[i])-55
    else:
      v[i] = int(v[i])
  return v

"""transformando os números na base 10 em símbolos equivalentes"""
def int_to_char(v): # vetor int --> vetor char
  for i in range(len(v)):
    if v[i] >=10:
      v[i] = chr(v[i]+55)
    else:
      v[i] = str(v[i])
  return v
```

```python
"""convertendo um número da base 10 para a base B"""
def b10_to_bB(x, b, casas_calculadas):
  x = float(x)
  inteiro, fracionario = int(x), x - int(x)
  # Convertendo a parte inteira
  resultado_int = []
  while inteiro > 0:
    resultado_int.insert(0, inteiro % b)
    inteiro //= b
  resultado_int = int_to_char(resultado_int) or ["0"]
  # Convertendo a parte fracionária
  resultado_frac = []
  for i in range(casas_calculadas):
    fracionario *= b
    digito = int(fracionario)
    resultado_frac.append(digito)
    fracionario -= digito
    if fracionario == 0:
      break
  resultado_frac = int_to_char(resultado_frac)
  # Construindo o resultado final
  if resultado_frac:
    return "".join(resultado_int) + "." + "".join(resultado_frac)
  else:
    return "".join(resultado_int)
```

```python
"""somar números da mesma base"""
def somar_mesma_base(x,y,b):
  x = list(x)
  y = list(y)
  #calculando quantas casas inteiras e decimais cada número tem
  if "." in x:
    casas_frac_x = len(x) - x.index(".") - 1
    x.remove(".")
  else:
    casas_frac_x = 0
  if "." in y:
    casas_frac_y = len(y) - y.index(".") - 1
    y.remove(".")
  else:
    casas_frac_y = 0
  casas_int_x = len(x)-casas_frac_x
  casas_int_y = len(y)-casas_frac_y
  # tranformando os símbolos em números equivalentes na base 10
  char_to_int(x)
  char_to_int(y)
  #colocando os vetores com o mesmo número de casas
  if casas_int_x>casas_int_y:
    for i in range(casas_int_x-casas_int_y):
      y.insert(0,0)
  else:
    for i in range(casas_int_y-casas_int_x):
      x.insert(0,0)
  if casas_frac_x>casas_frac_y:
    for i in range(casas_frac_x-casas_frac_y):
      y.append(0)
  else:
    for i in range(casas_frac_y-casas_frac_x):
      x.append(0)
  #adicionando mais uma casa para possível "vai um"
  x.insert(0,0)
  y.insert(0,0)
  #cálculo
  vai_um = 0
  resultado = []
  for i in range(len(x)-1,-1,-1):
    resultante = x[i]+y[i]+vai_um
    vai_um = 0
    if resultante>=b:
      vai_um+=1
      resultante-=b
    resultado.insert(0,resultante)
  #transformando números na base 10 para símbolos equivalentes
  int_to_char(resultado)
  #adicionando o "."
  resultado.insert(max(casas_int_x,casas_int_y)+1,".")
  #tirando possível 0 à esquerda (inteiros) e à direita (fracionados) e possível "."
  while resultado[0] == "0" and resultado[1] != ".":
    del resultado[0]
  while resultado[len(resultado)-1] == "0":
    del resultado[len(resultado)-1]
  if resultado[len(resultado)-1] == ".":
    del resultado[len(resultado)-1]
  #transformando resultado em string
  resultado = "".join(resultado)
  return resultado
```

```python
"""multiplicar dois números de bases iguais"""
def multiplicar_mesma_base(x,y,b):
  x = list(x)
  y = list(y)
  #calculando quantas casas decimais cada número tem
  if "." in x:
    casas_frac_x = len(x) - x.index(".") - 1
    x.remove(".")
  else:
    casas_frac_x = 0
  if "." in y:
    casas_frac_y = len(y) - y.index(".") - 1
    y.remove(".")
  else:
    casas_frac_y = 0
  # tranformando os símbolos em números equivalentes na base 10
  char_to_int(x)
  char_to_int(y)
  #adicionando mais uma casa para possível "vai um"
  x.insert(0,0)
  #cálculo multiplicação
  resultado = []
  for i in range(len(y)):
    resultado.append([])
    vai_um = 0
    for j in range(i):#botando casas necessarias nos vetores resultado[i]
      resultado[i].insert(0,0)
    for j in range(len(x)):
      produto = y[len(y)-1-i]*x[len(x)-1-j]+vai_um #int base 10
      resultado[i].insert(0,produto%b) #int
      vai_um = produto//b
    int_to_char(resultado[i]) #transformando int para char
    resultado[i] = "".join(resultado[i]) #transformando vetor char para str
  #cálculo soma
  soma = resultado[0]
  if len(resultado)>1:
    for i in range(len(resultado)-1):
      soma = somar_mesma_base(resultado[1+i],soma,b)
  #botando "."
  if (casas_frac_x+casas_frac_y) != 0:
    soma = list(soma)
    for i in range(1-(len(soma)-(casas_frac_x+casas_frac_y))):
      soma.insert(0,"0")
    soma.insert(len(soma)-(casas_frac_x+casas_frac_y),".")
    #tirando possível 0 à direita (fracionados) e possível "."
    while soma[len(soma)-1] == "0":
      del soma[len(soma)-1]
    if soma[len(soma)-1] == ".":
      del soma[len(soma)-1]
    soma = "".join(soma)
  return soma
```

```python
"""converter um número na base b1 para a base b2 diretamente (método de horner)"""
def mudar_base(x,b1,b2,casas_calculadas):
  casas_calculadas += 5 #margem de segurança para uma resposta precisa
  x = list(x)
  #calculando quantas casas decimais cada número tem
  if "." in x:
    casas_frac_x = len(x) - x.index(".") - 1
    x.remove(".")
  else:
    casas_frac_x = 0
  casas_int_x = len(x)-casas_frac_x
  #tranformando os símbolos em números equivalentes na base 10
  char_to_int(x)
  #cálculo
  base = b10_to_bB(b1,b2,casas_calculadas)
  inverso_base = b10_to_bB((1/b1),b2,casas_calculadas)
  resultado = b10_to_bB(x[0],b2,casas_calculadas)
  for i in range(len(x)-1):
    resultado = multiplicar_mesma_base(resultado,base,b2)
    resultado = somar_mesma_base(resultado,b10_to_bB(x[i+1],b2,casas_calculadas),b2)
  for i in range(casas_frac_x):
    resultado = multiplicar_mesma_base(resultado,inverso_base,b2)
  #tirando casas com algarismos duvidosos
  casas_calculadas -= 5 #tirando a margem
  resultado = list(resultado)
  if "." in resultado:
    casas_int_resultado = resultado.index(".")
  else:
    casas_int_resultado = len(resultado)
  while (casas_int_resultado+casas_calculadas+1) < len(resultado):
    del resultado[casas_int_resultado+casas_calculadas+1]
  resultado = "".join(resultado)
  #return
  return resultado
```

```python
"""somar dois números de bases diferentes"""
def somar_diferentes_bases(x,bx,y,by,bf,casas_calculadas): #(número 1, base do número 1, número 2, base do número 2, base final, precisão de casas)
  y = mudar_base(y,by,bx,casas_calculadas) #convertendo num 2 para a base 1
  resposta = somar_mesma_base(x,y,bx) #somando num 1 e num 2 na base 1
  resposta = mudar_base(resposta,bx,bf,casas_calculadas) #convertendo o resultado para a base final
  return resposta
"""multiplicar dois números de bases diferentes"""
def multiplicar_diferentes_bases(x,bx,y,by,bf,casas_calculadas): #(número 1, base do número 1, número 2, base do número 2, base final, precisão de casas)
  y = mudar_base(y,by,bx,casas_calculadas) #convertendo num 2 para a base 1
  resposta = multiplicar_mesma_base(x,y,bx) #multiplicando num 1 e num 2 na base 1
  resposta = mudar_base(resposta,bx,bf,casas_calculadas) #convertendo o resultado para a base final
  return resposta
```
